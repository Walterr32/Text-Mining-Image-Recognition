```python
def aplicar_color(globos, color):
    if globos is None: 
        raise ValueError("La imagen proporcionada es None")
    globos_salida = np.zeros_like(globos)
    if color == 1:
        globos_salida[:, :, 0] = globos[:, :, 0]
    elif color == 2:
        globos_salida[:, :, 1] = globos[:, :, 1]
    elif color == 3: 
        globos_salida[:, :, 2] = globos[:, :, 2]
    elif color == 10:
        globos_salida[:, :, 1] = globos[:, :, 1]
        globos_salida[:, :, 2] = globos[:, :, 2]
    elif color == 20:
        globos_salida[:, :, 1] = globos[:, :, 1]
        globos_salida[:, :, 0] = globos[:, :, 0]
    elif color == 30:
        globos_salida[:, :, 0] = globos[:, :, 0]
        globos_salida[:, :, 2] = globos[:, :, 2]
    else:
        raise ValueError("El valor de 'color' no es valido")
    return globos_salida
```

```python
color = 30
globos_resultado = aplicar_color(globos, color)
```

```python
plt.imshow(globos_resultado)
plt.show()
```

## Ejercicio 5

El enfoque de escala de grises ponderado consiste en convertir una imagen que esta a color a una escala de grises usando una combinacin pondera de los canales rojo, verde y azul. A diferencia del enfoque de aritmtico simple, mtodo ponderado toma en cuenta que el ojo humano no percibe los colores con la misma intensidad. El ojo humano es ms sensible al verde, luego rojo y azul.

Las ponderaciones ms usadas son:

Rojo: 0.299
Verde: 0.587
Azul: 0.114

```python
plt.imshow(fornite)
plt.axis("off")
plt.show()
```

```python
def weighted_grayscale(image):
    
    # Ponderaciones para los canales B, G, R
    weights = [0.114, 0.587, 0.299]
    
    # Convertir la imagen a flotante para evitar recortes durante la multiplicacin
    image = image.astype(float)
    
    # Aplicar la frmula ponderada
    gray_image = image[:,:,0] * weights[0] + image[:,:,1] * weights[1] + image[:,:,2] * weights[2]
    
    # Convertir de vuelta a entero de 8 bits
    gray_image = gray_image.astype(np.uint8)
    
    return gray_image
```

```python
gray_image = weighted_grayscale(fornite)
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(fornite, cv2.COLOR_BGR2RGB))
plt.title('Imagen en Color')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(gray_image, cmap='gray')
plt.title('Imagen en Escala de Grises Ponderada')
plt.axis('off')

plt.show()
```

## Ejercicio 6

El espacio de color HSV (Hue, Saturation, Value) es un modelo de color que representa colores de una manera que es ms intuitiva para los humanos en comparacin con el espacio de color RGB. A diferencia del modelo RGB, que se basa en la mezcla de colores primarios de luz (rojo, verde y azul)

Hue (Matiz): Representa el tipo de color y se mide en grados de un crculo (0 a 360)

Saturation (Saturacin): Indica la intensidad o pureza del color y se mide en porcentaje (0% a 100%). Un valor de 0% corresponde a un color completamente desaturado (gris), mientras que un valor de 100% corresponde a un color completamente saturado (vivo).

Value (Valor o Brillo): Representa el brillo del color y tambin se mide en porcentaje (0% a 100%). Un valor de 0% es completamente negro, mientras que un valor de 100% corresponde al color en su mxima intensidad.

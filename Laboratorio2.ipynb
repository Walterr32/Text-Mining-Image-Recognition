# Tranfosmación negativa

Se utilizara una imagen ya cargada, en este caso, somrbillas_gray

```python
plt.subplot(1, 2, 2)
plt.title('Sombrillas en Escala de Grises')
plt.imshow(sombrillas_gray, cmap='gray')
plt.axis('off')

plt.show()
```

# Transformación negativa

```python
def transformacion_negativa(imagen_gray):
    imagen_negativa = 255 - imagen_gray
    return imagen_negativa

imagen_negativa = transformacion_negativa(sombrillas_gray)
plt.imshow(imagen_negativa, cmap='gray')
plt.title('Transformación Negativa')
plt.axis('off')
plt.show
```

# Transformación Lineal

```python
def transformacion_lineal(imagen_gray, a = 1.0, b=0):
    imagen_lineal = cv2.convertScaleAbs(imagen_gray, alpha=a, beta=b)
    return imagen_lineal

imagen_lineal = transformacion_lineal(sombrillas_gray, a=1.5, b=50)

plt.imshow(imagen_lineal, cmap='gray')
plt.title('Transformación Lineal')
plt.axis('off')
plt.show
```

# Transformación Logaritmica

```python
def transformacion_logaritmica(imagen_gray, c=1):
    imagen_log = c * np.log(1 + imagen_gray.astype(np.float32))
    imagen_log = np.uint8(cv2.normalize(imagen_log, None, 0, 255, cv2.NORM_MINMAX))
    return imagen_log

imagen_logaritmica = transformacion_logaritmica(sombrillas_gray, c = 40)
plt.imshow(imagen_logaritmica, cmap='gray')
plt.title('Transformación Logaritmica')
plt.axis('off')
plt.show
```

# Transformación exponencial

```python
def transformacion_exponencial(imagen_gray, c=1):
    imagen_exp = c * (np.exp(imagen_gray.astype(np.float32)) -1)
    imagen_exp = np.uint8(cv2.normalize(imagen_exp, None, 0, 255, cv2.NORM_MINMAX))
    return imagen_exp

imagen_exponencial = transformacion_exponencial(sombrillas_gray, c=0.01)
plt.imshow(imagen_exponencial, cmap='gray')
plt.title('Transformación Exponencial')
plt.axis('off')
plt.show
```

## Ejercicio 5

El kernel laplaciano es un operador matemático utilizado en procesamiento de imágenes para detectar bordes. Este enfoque se basa en el operador Laplaciano, que es un segundo derivado, y mide la tasa de cambio en las intensidades de píxeles. A diferencia de otros métodos que detectan bordes buscando máximos en la primera derivada de la intensidad, el laplaciano detecta puntos donde la segunda derivada es cero, es decir, en puntos de inflexión donde hay un cambio brusco en la dirección de la intensidad.

# Características del Kernel Laplaciano:
Detección de Bordes en Todas las Direcciones: El kernel Laplaciano detecta bordes en todas las direcciones, ya que considera la segunda derivada de la imagen.
Sensibilidad al Ruido: Debido a que el operador laplaciano calcula la segunda derivada, puede ser muy sensible al ruido en la imagen. Un pequeño cambio en la intensidad puede causar un gran cambio en la segunda derivada.
Simplicidad: El operador laplaciano es simple de implementar, pero no tiene una capacidad intrínseca para distinguir entre bordes significativos y ruido, lo que puede llevar a la detección de bordes falsos.

# Diferencias entre Kernel Laplaciano y Algoritmo de Canny
Metodología:

Kernel Laplaciano: Utiliza la segunda derivada de la intensidad para detectar bordes. Es una técnica más simple, pero menos robusta frente al ruido.
Canny: Utiliza la primera derivada con una serie de pasos adicionales, como suavizado y supresión de no máximos, lo que lo hace más preciso y robusto.
Sensibilidad al Ruido:

Kernel Laplaciano: Es más sensible al ruido, ya que no tiene una etapa de suavizado previa.
Canny: Incluye un filtrado gaussiano inicial para reducir el ruido, lo que lo hace más efectivo en imágenes ruidosas.
Control sobre la Detección:

Kernel Laplaciano: Proporciona menos control, ya que no incluye parámetros de umbral.
Canny: Ofrece un mayor control a través de los umbrales de histéresis, permitiendo ajustar la sensibilidad en la detección de bordes.
Precisión:

Kernel Laplaciano: Tiende a detectar más bordes falsos debido a su sensibilidad.
Canny: Es más preciso en la detección de bordes verdaderos y reduce la probabilidad de detecciones falsas.

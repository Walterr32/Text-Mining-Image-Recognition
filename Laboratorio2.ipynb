# Tranfosmacin negativa

Se utilizara una imagen ya cargada, en este caso, somrbillas_gray

```python
plt.subplot(1, 2, 2)
plt.title('Sombrillas en Escala de Grises')
plt.imshow(sombrillas_gray, cmap='gray')
plt.axis('off')

plt.show()
```

# Transformacin negativa

```python
def transformacion_negativa(imagen_gray):
    imagen_negativa = 255 - imagen_gray
    return imagen_negativa

imagen_negativa = transformacion_negativa(sombrillas_gray)
plt.imshow(imagen_negativa, cmap='gray')
plt.title('Transformacin Negativa')
plt.axis('off')
plt.show
```

# Transformacin Lineal

```python
def transformacion_lineal(imagen_gray, a = 1.0, b=0):
    imagen_lineal = cv2.convertScaleAbs(imagen_gray, alpha=a, beta=b)
    return imagen_lineal

imagen_lineal = transformacion_lineal(sombrillas_gray, a=1.5, b=50)

plt.imshow(imagen_lineal, cmap='gray')
plt.title('Transformacin Lineal')
plt.axis('off')
plt.show
```

# Transformacin Logaritmica

```python
def transformacion_logaritmica(imagen_gray, c=1):
    imagen_log = c * np.log(1 + imagen_gray.astype(np.float32))
    imagen_log = np.uint8(cv2.normalize(imagen_log, None, 0, 255, cv2.NORM_MINMAX))
    return imagen_log

imagen_logaritmica = transformacion_logaritmica(sombrillas_gray, c = 40)
plt.imshow(imagen_logaritmica, cmap='gray')
plt.title('Transformacin Logaritmica')
plt.axis('off')
plt.show
```

# Transformacin exponencial

```python
def transformacion_exponencial(imagen_gray, c=1):
    imagen_exp = c * (np.exp(imagen_gray.astype(np.float32)) -1)
    imagen_exp = np.uint8(cv2.normalize(imagen_exp, None, 0, 255, cv2.NORM_MINMAX))
    return imagen_exp

imagen_exponencial = transformacion_exponencial(sombrillas_gray, c=0.01)
plt.imshow(imagen_exponencial, cmap='gray')
plt.title('Transformacin Exponencial')
plt.axis('off')
plt.show
```

## Ejercicio 5

El kernel laplaciano es un operador matemtico utilizado en procesamiento de imgenes para detectar bordes. Este enfoque se basa en el operador Laplaciano, que es un segundo derivado, y mide la tasa de cambio en las intensidades de pxeles. A diferencia de otros mtodos que detectan bordes buscando mximos en la primera derivada de la intensidad, el laplaciano detecta puntos donde la segunda derivada es cero, es decir, en puntos de inflexin donde hay un cambio brusco en la direccin de la intensidad.

# Caractersticas del Kernel Laplaciano:
Deteccin de Bordes en Todas las Direcciones: El kernel Laplaciano detecta bordes en todas las direcciones, ya que considera la segunda derivada de la imagen.
Sensibilidad al Ruido: Debido a que el operador laplaciano calcula la segunda derivada, puede ser muy sensible al ruido en la imagen. Un pequeo cambio en la intensidad puede causar un gran cambio en la segunda derivada.
Simplicidad: El operador laplaciano es simple de implementar, pero no tiene una capacidad intrnseca para distinguir entre bordes significativos y ruido, lo que puede llevar a la deteccin de bordes falsos.

# Diferencias entre Kernel Laplaciano y Algoritmo de Canny
Metodologa:

Kernel Laplaciano: Utiliza la segunda derivada de la intensidad para detectar bordes. Es una tcnica ms simple, pero menos robusta frente al ruido.
Canny: Utiliza la primera derivada con una serie de pasos adicionales, como suavizado y supresin de no mximos, lo que lo hace ms preciso y robusto.
Sensibilidad al Ruido:

Kernel Laplaciano: Es ms sensible al ruido, ya que no tiene una etapa de suavizado previa.
Canny: Incluye un filtrado gaussiano inicial para reducir el ruido, lo que lo hace ms efectivo en imgenes ruidosas.
Control sobre la Deteccin:

Kernel Laplaciano: Proporciona menos control, ya que no incluye parmetros de umbral.
Canny: Ofrece un mayor control a travs de los umbrales de histresis, permitiendo ajustar la sensibilidad en la deteccin de bordes.
Precisin:

Kernel Laplaciano: Tiende a detectar ms bordes falsos debido a su sensibilidad.
Canny: Es ms preciso en la deteccin de bordes verdaderos y reduce la probabilidad de detecciones falsas.
